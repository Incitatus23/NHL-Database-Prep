```python
import pandas as pd
import numpy as np
```

```python
teamdf = pd.read_csv('TeamInfo.csv')
gamedf = pd.read_csv('GameData.csv')
```

```python
teamdf.head()
```

```python
gamedf.info()
```

```python
game_team = pd.merge(teamdf, gamedf, how="inner", left_on="team_id", right_on="home_team_id")
```

```python
game_team
```

```python
gameteam2 = pd.merge(teamdf, game_team, how="inner", left_on="team_id", right_on="away_team_id")
```

```python
gameteam2
```

```python
gameteam2.info()
```

```python
gameteam3 = gameteam2[['teamName_x', 'teamName_y', 'date_time_GMT', 'away_goals', 'home_goals']]
```

```python
gameteam3
```

```python
gameteam4 = gameteam3.rename(columns={'teamName_x': 'homemascot', 'teamName_y': 'awaymascot', 'away_goals':'awaymascotscore',
                                     'home_goals':'homemascotscore'})
```

```python
gameteam4
```

```python
gameteam4['date_time_GMT'] =  pd.to_datetime(gameteam4['date_time_GMT'])
```

```python
gameteam4['gamedate'] = gameteam4['date_time_GMT'].dt.date
```

```python
gameteam4 =  gameteam4.drop(columns=['date_time_GMT'])
```

```python
gameteam4
```

```python
gameteam4.to_csv('gameinfo.csv')
```
